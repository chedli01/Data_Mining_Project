```python
# Valeurs manquantes
missing_data = pd.DataFrame({
    'Colonne': df.columns,
    'Valeurs_Manquantes': df.isnull().sum(),
    'Pourcentage': (df.isnull().sum() / len(df) * 100).round(2)
})
missing_data = missing_data[missing_data['Valeurs_Manquantes'] > 0].sort_values(
    'Valeurs_Manquantes', ascending=False
)

print("ANALYSE DES VALEURS MANQUANTES")
print("="*60)
print(missing_data)

# Visualisation
if len(missing_data) > 0:
    plt.figure(figsize=(12, 6))
    plt.barh(missing_data['Colonne'], missing_data['Pourcentage'])
    plt.xlabel('Pourcentage de valeurs manquantes')
    plt.title('Valeurs manquantes par colonne')
    plt.tight_layout()
    plt.show()
else:
    print("\n✅ Aucune valeur manquante détectée!")
```

```python
# Sélection des colonnes utiles (disponibles AVANT la sortie du film)
cols = [
    "budget",
    "runtime",
    "genres",
    "popularity",
    "original_language",
    "release_date",  # Ajouté pour analyse temporelle
    "vote_average",   # NOTE: Peut être data leakage - à analyser
    "vote_count",     # NOTE: Peut être data leakage - à analyser
    "revenue"
]
   
df = df[cols].copy()

print(f"✅ Dataset réduit à {df.shape[1]} colonnes pertinentes")
print(f"Colonnes sélectionnées: {df.columns.tolist()}")

df.head()
```

```python
print("DIAGNOSTIC QUALITÉ DES DONNÉES (AVANT NETTOYAGE)")
print("=" * 60)

# Taille initiale
print(f"Taille initiale du dataset : {df.shape}")

# 1. Valeurs manquantes (NaN)
print("\nVALEURS MANQUANTES (NaN)")
missing = df.isna().sum()
missing_percent = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    "Valeurs manquantes": missing,
    "Pourcentage (%)": missing_percent.round(2)
})

print(missing_df[missing_df["Valeurs manquantes"] > 0])

# 2. Valeurs nulles (0) pour variables numériques
print("\nVALEURS NULLES (0) DANS LES VARIABLES NUMÉRIQUES")

numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

zeros_df = pd.DataFrame({
    "Valeurs nulles (0)": (df[numeric_cols] == 0).sum(),
    "Pourcentage (%)": ((df[numeric_cols] == 0).sum() / len(df) * 100).round(2)
})

print(zeros_df[zeros_df["Valeurs nulles (0)"] > 0])

# 3. Valeurs négatives (anomalies)
print("\nVALEURS NÉGATIVES (ANOMALIES POTENTIELLES)")

negatives = {
    col: (df[col] < 0).sum()
    for col in numeric_cols
}

neg_df = pd.DataFrame.from_dict(
    negatives, orient="index", columns=["Nombre de valeurs négatives"]
)

print(neg_df[neg_df["Nombre de valeurs négatives"] > 0])

# 4. Résumé statistique des variables numériques
print("\nRÉSUMÉ STATISTIQUE DES VARIABLES NUMÉRIQUES")
display(df[numeric_cols].describe())

print("=" * 60)
print("Fin du diagnostic")
print("=" * 60)
```

```python
print("NETTOYAGE DES DONNÉES (PIPELINE STRUCTURÉ)")
print("="*60)

# 1. Supprimer NaN critiques
df = df.dropna(subset=["runtime", "release_date"])

# 2. Runtime invalide
df = df[df["runtime"] > 0]

# 3. Budget
df["budget"] = pd.to_numeric(df["budget"], errors="coerce")
df["budget_missing"] = (df["budget"] == 0).astype(int)
median_budget = df.loc[df["budget"] > 0, "budget"].median()
df.loc[df["budget"] <= 0, "budget"] = median_budget

# 4. Revenue
df["revenue"] = pd.to_numeric(df["revenue"], errors="coerce")
df = df[df["revenue"] >= 0]

# 5. Reset index
df.reset_index(drop=True, inplace=True)

# 6. Vérification
print("\nVérification post-nettoyage:")
print("Budget <= 0 :", (df["budget"] <= 0).sum())
print("Runtime <= 0:", (df["runtime"] <= 0).sum())
print("Revenue < 0 :", (df["revenue"] < 0).sum())
print("NaN restant :", df.isna().sum().sum())
```

```python
# Fonction pour extraire les noms de genres
def extract_genre_names(genres_str):
    """
    Extrait les noms de genres depuis la chaîne JSON
    Exemple: '[{"id": 28, "name": "Action"}]' -> ['Action']
    """
    try:
        genres = ast.literal_eval(genres_str)
        return [g["name"] for g in genres]
    except:
        return []

# Application de la fonction
df["genres"] = df["genres"].apply(extract_genre_names)

print("EXTRACTION DES GENRES")
print("="*60)

# Analyse des genres
all_genres = [genre for sublist in df["genres"] for genre in sublist]
genre_counts = pd.Series(all_genres).value_counts()

print(f"Nombre total de genres uniques: {len(genre_counts)}")
print(f"\nTop 10 des genres les plus fréquents:")
print(genre_counts.head(10))

# Visualisation
plt.figure(figsize=(12, 6))
genre_counts.head(15).plot(kind='barh')
plt.xlabel('Nombre de films')
plt.title('Top 15 des genres les plus fréquents')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

print("CRÉATION DE LA VARIABLE CIBLE: SUCCESS (3 CLASSES)")
print("="*60)

# --- 0. Assurer que budget est numérique ---
df["budget"] = pd.to_numeric(df["budget"], errors="coerce")

# --- 1. Indicateur de budget manquant ---
df["budget_missing"] = df["budget"].isna().astype(int)

# --- 2. Imputation du budget par la médiane ---
median_budget = df["budget"].median()
df["budget"].fillna(median_budget, inplace=True)
print(f"Budgets manquants imputés par la médiane : {median_budget:.2f}")

# --- 3. Calcul du ROI ---
df["ROI"] = df["revenue"] / df["budget"]

# --- 4. Définition de la variable cible "success" en 3 classes ---
# 0 = échec (ROI < 1), 1 = succès modéré (1 <= ROI < 2), 2 = grand succès (ROI >= 2)
df["success"] = pd.cut(
    df["ROI"],
    bins=[-np.inf, 1, 2, np.inf],
    labels=[0, 1, 2]
).astype(int)

# --- 5. Distribution et proportions ---
print("\nDistribution de la variable cible:")
print(df["success"].value_counts())
print("\nProportion:")
print(df["success"].value_counts(normalize=True).round(3))

# --- 6. Visualisation ---
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Count plot
sns.countplot(x="success", data=df, ax=axes[0], palette=['#ff6b6b', '#ffd43b', '#51cf66'])
axes[0].set_title('Distribution de la variable cible', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Succès (0=Échec, 1=Succès modéré, 2=Grand succès)')
axes[0].set_ylabel('Nombre de films')
axes[0].set_xticklabels(['Échec', 'Succès modéré', 'Grand succès'])

# Pie chart
colors = ['#ff6b6b', '#ffd43b', '#51cf66']
df["success"].value_counts().sort_index().plot(
    kind='pie',
    ax=axes[1],
    autopct='%1.1f%%',
    colors=colors,
    labels=['Échec', 'Succès modéré', 'Grand succès']
)
axes[1].set_title('Répartition Succès/Échec', fontsize=14, fontweight='bold')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
print("ANALYSE: BUDGET vs SUCCÈS (3 CLASSES)")
print("="*60)

# Statistiques par catégorie
budget_stats = df.groupby('success')['budget'].describe()
print("\nStatistiques du budget par catégorie:")
print(budget_stats)

# Visualisations
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. Distribution du budget
axes[0, 0].hist(df["budget"], bins=50, edgecolor='black', alpha=0.7)
axes[0, 0].set_title('Distribution du Budget (tous films)', fontweight='bold')
axes[0, 0].set_xlabel('Budget ($)')
axes[0, 0].set_ylabel('Fréquence')
axes[0, 0].axvline(df["budget"].median(), color='red', linestyle='--', 
                    label=f'Médiane: ${df["budget"].median():,.0f}')
axes[0, 0].legend()

# 2. Budget par succès (boxplot)
palette_colors = ['#ff6b6b', '#ffd43b', '#51cf66']
sns.boxplot(x='success', y='budget', data=df, ax=axes[0, 1], palette=palette_colors)
axes[0, 1].set_title('Budget par catégorie de succès', fontweight='bold')
axes[0, 1].set_xticklabels(['Échec', 'Succès modéré', 'Grand succès'])
axes[0, 1].set_ylabel('Budget ($)')

# 3. Distribution du budget (log scale) par succès
for cls, color, label in zip([0, 1, 2], palette_colors, ['Échec', 'Succès modéré', 'Grand succès']):
    axes[1, 0].hist(df[df['success'] == cls]['budget'], bins=50, alpha=0.5, label=label, color=color)
axes[1, 0].set_title('Distribution du budget par succès', fontweight='bold')
axes[1, 0].set_xlabel('Budget ($)')
axes[1, 0].set_ylabel('Fréquence')
axes[1, 0].set_yscale('log')
axes[1, 0].legend()

# 4. Taux de succès par tranche de budget
df['budget_category'] = pd.qcut(df['budget'], q=5, labels=['Très faible', 'Faible', 'Moyen', 'Élevé', 'Très élevé'],    duplicates='drop' )
success_by_budget = df.groupby('budget_category')['success'].value_counts(normalize=True).unstack(fill_value=0) * 100

success_by_budget.plot(kind='bar', stacked=True, ax=axes[1, 1], color=palette_colors, edgecolor='black')
axes[1, 1].set_title('Répartition des classes de succès par tranche de budget', fontweight='bold')
axes[1, 1].set_xlabel('Catégorie de budget')
axes[1, 1].set_ylabel('Pourcentage (%)')
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].legend(['Échec', 'Succès modéré', 'Grand succès'], title='Classe de succès')

plt.tight_layout()
plt.show()

# Supprimer la colonne temporaire
df.drop('budget_category', axis=1, inplace=True)
```

```python
print("ANALYSE: GENRES vs SUCCÈS (3 CLASSES)")
print("="*60)

# Calculer le nombre de films et le succès par genre et par classe
genre_success = {}
for idx, row in df.iterrows():
    for genre in row['genres']:
        if genre not in genre_success:
            genre_success[genre] = {0:0, 1:0, 2:0, 'total':0}
        genre_success[genre][row['success']] += 1
        genre_success[genre]['total'] += 1

# Créer un DataFrame
genre_df = pd.DataFrame([
    {
        'Genre': genre,
        'Total Films': data['total'],
        'Échec': data[0],
        'Succès modéré': data[1],
        'Grand succès': data[2],
        'Taux Succès modéré (%)': (data[1]/data['total']*100),
        'Taux Grand succès (%)': (data[2]/data['total']*100)
    }
    for genre, data in genre_success.items()
])

# Trier par taux de Grand succès
genre_df = genre_df.sort_values('Taux Grand succès (%)', ascending=False)

print("\nTaux de succès par genre:")
print(genre_df[['Genre','Total Films','Échec','Succès modéré','Grand succès']].to_string(index=False))

# Visualisations
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Taux de Grand succès par genre (min 50 films)
genre_df_filtered = genre_df[genre_df['Total Films'] >= 50]
genre_df_filtered.set_index('Genre')['Taux Grand succès (%)'].sort_values().plot(
    kind='barh', ax=axes[0, 0], color='#e67e22', edgecolor='black'
)
axes[0, 0].set_title('Taux de Grand succès par genre (min 50 films)', fontweight='bold')
axes[0, 0].set_xlabel('Taux de Grand succès (%)')
axes[0, 0].axvline(x=genre_df['Grand succès'].sum()/genre_df['Total Films'].sum()*100, 
                   color='red', linestyle='--', label='Moyenne globale')
axes[0, 0].legend()

# 2. Nombre de films par genre (Top 15)
genre_df.set_index('Genre')['Total Films'].sort_values(ascending=False).head(15).plot(
    kind='barh', ax=axes[0, 1], color='#3498db', edgecolor='black'
)
axes[0, 1].set_title('Top 15 genres par nombre de films', fontweight='bold')
axes[0, 1].set_xlabel('Nombre de films')
axes[0, 1].invert_yaxis()

# 3. Popularité du genre vs Taux de Grand succès
axes[1, 0].scatter(genre_df['Total Films'], genre_df['Taux Grand succès (%)'], 
                   s=100, alpha=0.6, edgecolors='black', linewidth=1)
axes[1, 0].set_title('Popularité du genre vs Taux de Grand succès', fontweight='bold')
axes[1, 0].set_xlabel('Nombre total de films')
axes[1, 0].set_ylabel('Taux de Grand succès (%)')
axes[1, 0].axhline(y=genre_df['Grand succès'].sum()/genre_df['Total Films'].sum()*100, 
                   color='red', linestyle='--', alpha=0.5)
axes[1, 0].set_xscale('log')

# Annoter genres extrêmes
for idx, row in genre_df.iterrows():
    if row['Total Films'] > 500 or row['Taux Grand succès (%)'] > 75 or row['Taux Grand succès (%)'] < 30:
        axes[1, 0].annotate(row['Genre'], 
                            (row['Total Films'], row['Taux Grand succès (%)']),
                            fontsize=8, alpha=0.7)

# 4. Nombre moyen de genres par film et succès
df['num_genres'] = df['genres'].apply(len)
success_by_num_genres = df.groupby('num_genres')['success'].value_counts(normalize=True).unstack(fill_value=0)

# Tracer le taux de chaque classe
success_by_num_genres.plot(kind='bar', stacked=True, ax=axes[1, 1], 
                           color=['#ff6b6b', '#ffd43b', '#51cf66'], edgecolor='black')
axes[1, 1].set_title('Répartition des classes de succès par nombre de genres', fontweight='bold')
axes[1, 1].set_xlabel('Nombre de genres par film')
axes[1, 1].set_ylabel('Proportion (%)')
axes[1, 1].tick_params(axis='x', rotation=0)
axes[1, 1].legend(['Échec', 'Succès modéré', 'Grand succès'], title='Classe de succès')

plt.tight_layout()
plt.show()
```

```python
def evaluate_model(model, X_test, y_test, model_name):
    # Prédictions
    y_pred = model.predict(X_test)
    
    # Calcul des métriques
    metrics = {
        "Model": model_name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average='weighted', zero_division=0),
        "Recall": recall_score(y_test, y_pred, average='weighted', zero_division=0),
        "F1-score": f1_score(y_test, y_pred, average='weighted', zero_division=0)
    }
    
    # Matrice de confusion
    cm = confusion_matrix(y_test, y_pred)
    
    # Affichage des résultats
    print("\n" + "="*60)
    print(f"📊 Évaluation du modèle: {model_name}")
    print("="*60)
    print(f"Accuracy : {metrics['Accuracy']:.4f}")
    print(f"Precision: {metrics['Precision']:.4f}")
    print(f"Recall   : {metrics['Recall']:.4f}")
    print(f"F1-score : {metrics['F1-score']:.4f}")
    
    print("\nMatrice de confusion:")
    print(cm)
    
    # Visualisation de la matrice de confusion
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Matrice de confusion - {model_name}')
    plt.xlabel('Prédit')
    plt.ylabel('Réel')
    plt.show()
    
    return metrics, cm
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

print("\n" + "📊 MODÈLE 3: NAÏVE BAYES")
print("="*60)

# 1️⃣ Imputer les NaN avec la médiane
imputer = SimpleImputer(strategy='median')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# 2️⃣ Standardiser les features (optionnel pour Naive Bayes, mais aide parfois)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)

# 3️⃣ Entraînement
nb = GaussianNB()
nb.fit(X_train_scaled, y_train)

# 4️⃣ Évaluation
nb_metrics, nb_cm = evaluate_model(nb, X_test_scaled, y_test, "Naive Bayes")

# 5️⃣ Sauvegarder les résultats
nb_results = nb_metrics.copy()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

print("\n" + "🎒 MODÈLE 4: BAGGING")
print("="*60)

# 1️⃣ Imputer les NaN
imputer = SimpleImputer(strategy='median')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# 2️⃣ Standardiser les features (optionnel pour Bagging/DecisionTree)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)

# 3️⃣ Entraînement
bag = BaggingClassifier(
    estimator=DecisionTreeClassifier(max_depth=10),
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)
bag.fit(X_train_scaled, y_train)

# 4️⃣ Évaluation
bag_metrics, bag_cm = evaluate_model(bag, X_test_scaled, y_test, "Bagging")

# 5️⃣ Sauvegarder les résultats
bag_results = bag_metrics.copy()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

print("\n" + "🌲 MODÈLE 5: RANDOM FOREST")
print("="*60)

# 1️⃣ Imputer les NaN (médiane)
imputer = SimpleImputer(strategy='median')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# 2️⃣ Standardisation (optionnelle pour RandomForest, mais ok à faire)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)

# 3️⃣ Entraînement
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=15,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train_scaled, y_train)

# 4️⃣ Évaluation
rf_metrics, rf_cm = evaluate_model(rf, X_test_scaled, y_test, "Random Forest")

# 5️⃣ Sauvegarder les résultats
rf_results = rf_metrics.copy()
```

```python

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

print("\n" + "🚀 MODÈLE 7: RÉGRESSION LOGISTIQUE")
print("="*60)

# 1️⃣ Imputer les NaN (médiane)
imputer = SimpleImputer(strategy='median')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# 2️⃣ Standardisation (très important pour Logistic Regression)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)

# 3️⃣ Entraînement
lr = LogisticRegression(max_iter=1000, random_state=42, n_jobs=-1)
lr.fit(X_train_scaled, y_train)

# 4️⃣ Évaluation
lr_metrics, lr_cm = evaluate_model(lr, X_test_scaled, y_test, "Logistic Regression")

# 5️⃣ Sauvegarder les résultats
lr_results = lr_metrics.copy()
```

```python
print("\n" + "📊 COMPARAISON DE TOUS LES MODÈLES")
print("="*60)

# Créer un DataFrame de comparaison (les modèles doivent déjà être entraînés sur X_train / y_train)
results_comparison = pd.DataFrame([
    dt_results,
    knn_results,
    nb_results,
    bag_results,
    rf_results,
    lr_results
])

# Trier par F1-score
results_comparison = results_comparison.sort_values('F1-score', ascending=False)

print("\n" + results_comparison.to_string(index=False))

# Visualisation comparative
metrics_to_plot = ['Accuracy', 'Precision', 'Recall', 'F1-score']

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

for idx, metric in enumerate(metrics_to_plot):
    ax = axes[idx // 2, idx % 2]
    
    results_sorted = results_comparison.sort_values(metric, ascending=True)
    
    ax.barh(results_sorted['Model'], results_sorted[metric], 
            color='#3498db', edgecolor='black')
    ax.set_xlabel(metric)
    ax.set_title(f'Comparaison des modèles - {metric}', fontweight='bold')
    ax.set_xlim([0, 1])
    
    # Ajouter les valeurs à côté des barres
    for i, v in enumerate(results_sorted[metric]):
        ax.text(v + 0.01, i, f'{v:.3f}', va='center')

plt.suptitle('Comparaison des Performances des Modèles', 
             fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Identifier le meilleur modèle
best_model_name = results_comparison.iloc[0]['Model']
best_f1 = results_comparison.iloc[0]['F1-score']

print(f"\n{'='*60}")
print(f"🏆 MEILLEUR MODÈLE: {best_model_name}")
print(f"📊 F1-Score: {best_f1:.4f}")
print(f"{'='*60}")
```

```python
print("\n" + "🎯 IMPORTANCE DES VARIABLES (FEATURE IMPORTANCE)")
print("="*60)

# Vérifier que best_rf est défini
if 'best_rf' not in locals():
    best_rf = rf  # Utiliser ton RandomForest par défaut si best_rf non défini

# Récupérer les importances
importances = best_rf.feature_importances_

# Créer un DataFrame
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
}).sort_values('Importance', ascending=False)

print("\nTop 20 des features les plus importantes:")
print(feature_importance_df.head(20).to_string(index=False))

# Visualisation
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Top 20 features (barh)
top_20 = feature_importance_df.head(20)
axes[0].barh(range(len(top_20)), top_20['Importance'], color='#e74c3c', edgecolor='black')
axes[0].set_yticks(range(len(top_20)))
axes[0].set_yticklabels(top_20['Feature'])
axes[0].set_xlabel('Importance')
axes[0].set_title('Top 20 Features les Plus Importantes', fontsize=14, fontweight='bold')
axes[0].invert_yaxis()

# Top 10 features (pie chart)
top_10 = feature_importance_df.head(10)
others_importance = feature_importance_df.iloc[10:]['Importance'].sum()
pie_data = list(top_10['Importance']) + [others_importance]
pie_labels = list(top_10['Feature']) + ['Autres']

axes[1].pie(pie_data, labels=pie_labels, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Répartition de l\'Importance (Top 10)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# Analyse par catégorie
print("\n" + "="*60)
print("IMPORTANCE PAR CATÉGORIE")
print("="*60)

def categorize_feature(feature):
    if feature.startswith('genre_'):
        return 'Genre'
    elif feature.startswith('original_language_'):
        return 'Langue'
    elif feature.startswith('era_'):
        return 'Ère'
    elif feature in ['budget', 'runtime', 'popularity', 'vote_average', 'vote_count', 
                     'release_year', 'num_genres', 'is_english']:
        return 'Numérique'
    else:
        return 'Autre'

feature_importance_df['Category'] = feature_importance_df['Feature'].apply(categorize_feature)
category_importance = feature_importance_df.groupby('Category')['Importance'].sum().sort_values(ascending=False)

print("\nImportance totale par catégorie:")
print(category_importance)

# Visualisation
plt.figure(figsize=(10, 6))
category_importance.plot(kind='bar', color='#3498db', edgecolor='black')
plt.title('Importance Totale par Catégorie de Features', fontsize=14, fontweight='bold')
plt.xlabel('Catégorie')
plt.ylabel('Importance Totale')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
